```python
def powlaw(x, a, b) :
    return a * np.power(x, -np.abs(b))

def powlaw_weird(x, a, b) :
    return a * np.power(x, np.maximum(-5.0,(-np.abs(b)-0.3)))

def powtan(x,a,b):
    return a* np.power(x, -np.abs(b)) * ((-np.arctan(x - 1.7 ))/np.pi + 1/2 )

def linlaw(x, a, b) :
    return (a*x+b) * np.exp(-x)

def polylaw(x,a,b,c) :
    return (a*x**2 + b*x +c) * np.exp(-10*x)

def curve_fit_log(xdata, ydata, method ) :
    
    #print xdata_log, ydata_log
    if method ==1:
        popt, pcov = curve_fit(powlaw_weird, xdata, ydata , maxfev = 10000)
    elif method ==2:
        popt, pcov = curve_fit(powtan, xdata, ydata , maxfev = 10000)
    return popt, pcov

    
```

```python
def just_fit(r, xi):
    try:
        popt,_ = curve_fit_log(r/60.0, xi)
    except RuntimeError:
        print("using zero fitting")
        popt,_ = curve_fit_log(r/60.0, np.zeros(20))
    a = popt[0]
    b = popt[1]
    

    #print(b)
    
    #c = popt[2]
    #print(c)    
    
    #print(a,b)
    #print a,b
    theta_list = np.arange(0.01,30.01,0.001)

    corr_list = []
    for theta in theta_list:
        corr_list.append(powlaw_weird(theta,a,b))
    theta_rad = theta_list/180.0*np.pi
    corr_list = np.array(corr_list)
    
    return theta_rad, corr_list
```

```python
def find_Cl(r, xi,method):
    
    
    try:
        popt,_ = curve_fit_log(r/60.0, xi, method)
    except RuntimeError:
        print("using zero fitting")
        popt,_ = curve_fit_log(r/60.0, np.zeros(20), method)
    a = popt[0]
    b = popt[1]
    #c = popt[2]
    
#     if method ==2:
#         c = popt[2]
    
    #print(c)
    #print a,b
    theta_list = np.arange(0.01,180,0.001)
    cosine_list = np.cos(theta_list/180.0*np.pi)

    corr_list = []
    for theta in theta_list:
        if method ==1:
            corr_list.append(powlaw_weird(theta,a,b))
        if method ==2:
            corr_list.append(powtan(theta,a,b))
    theta_rad = theta_list/180.0*np.pi
    corr_list = np.array(corr_list)
    weight = np.sin(np.array(theta_rad))*np.pi/corr_list.shape[0]

    
    corr_array = np.array([corr_list,corr_list,corr_list,corr_list]).T
    
    Cl_camb = correlations.corr2cl(corr_array, cosine_list , weight, ell_max)
    
    
    
    #return ell_list_camb, Cl_camb[:,0]
    return ell_list_camb, Cl_camb[:,0], popt

    
```

```python
def Cl2maps(Cls, NSIDE,nob):
    Cl_for_hp = []
    for i in range(nob):
        for j in range(i,nob):
            Cl_for_hp.append(Cls[i][j])
    Cl_for_hp = np.array(Cl_for_hp)
    Cl_cut_zero = Cl_for_hp[:,1:]
    ell_int = np.arange(1,3072+1,1)
    Cl_ready = np.divide(np.divide(Cl_cut_zero,ell_int),(ell_int+1))*2 * np.pi

    alms = hp.sphtfunc.synalm(Cl_ready)
    maps = hp.sphtfunc.alm2map(alms, NSIDE,pol = False)
    return maps
```

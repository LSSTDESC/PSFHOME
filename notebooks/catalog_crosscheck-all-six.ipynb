```python
catalog = pd.read_csv("172382.csv",header = 0)
```

```python
ra_array, dec_array = catalog['ra'],catalog['dec']
```

```python
position_matrix = np.array([ra_array,dec_array]).T
```

```python
print catalog.shape
```

```python
import scipy
```

```python
kdtree = scipy.spatial.KDTree(position_matrix)
```

```python
#minimum 2 arcsec separation: 
min_sep = 2.0/3600
query_pair = kdtree.query_pairs(min_sep)
#print query_pair
```

```python
catalog.head()
```

```python
print len(query_pair)
```

```python
i=0
for pair in query_pair:
    pos1 = [catalog['ra'][pair[0]],catalog['dec'][pair[0]] ]
    pos2 = [catalog['ra'][pair[1]],catalog['dec'][pair[1]] ]
    
    dis = np.sqrt((pos2[0]-pos1[0])**2+(pos2[1]-pos1[1])**2)
    print pos1
    print pos2
    print dis
    print '\n'
    
    i+=1
    if i==10:
        break
```

```python
catalog['near_neighbor'] = False
for pair in query_pair:
    catalog.at[pair[0],'near_neighbor'] = True
    catalog.at[pair[1],'near_neighbor'] = True
```

```python
all_s16_wide_bad = catalog[catalog['near_neighbor']==True]
```

```python
all_s16_wide_bad.to_csv(r'all_s16_wide_bad.csv')
```

```python
id_set = set()
i=0
for item in query_pair:
    id1 = catalog["# object_id"][item[0]]
    id2 = catalog["# object_id"][item[1]]
    id_set.add(id1)
    id_set.add(id2)

    i=i+1
    
    if i%1000000==0:
        clear_output()
        print i
```

```python
highSNR_cata = pd.read_csv("GAMA15H_SNR1000_NOBLEND_NOSAT_NOSUSPECT_STAR_FILEMATCHING.csv",header = 0)
highSNR_cata['near_neighbor'] = False


for i in range(len(highSNR_cata['# object_id'])):
    if highSNR_cata['# object_id'][i] in id_set:
        highSNR_cata.at[i,'near_neighbor'] = True


```
